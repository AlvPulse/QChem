```python
from scipy.special import comb
p=0.95
k= 95
n=100
ProbXK= comb(n,k,exact=True)*p**k*(1-p)**(n-k)
ProbXK

```

```python
from scipy.stats import binom
CDF= binom.cdf(k, n, p)
print(CDF)
```

```python
import matplotlib.pyplot as plt
import numpy as np
pdf= np.zeros((101,))
cdf= np.zeros((101,))
for x in range(0,101):
    #pdf[x]= comb(n,x,exact=True)*p**x*(1-p)**(n-x)
    #print(x)
    pdf[x]= binom.pmf(x,n,p)
    cdf[x]= binom.cdf(x,n, p)
x_array= np.arange(0,101)
print(pdf.shape)
print(x_array.shape)
plt.plot(x_array, pdf)
plt.plot(x_array, cdf)
plt.legend(["pdf","CDF"])
```
